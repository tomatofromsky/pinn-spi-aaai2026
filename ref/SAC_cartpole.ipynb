```python
#!/usr/bin/env python3
# ------------------------------------------------------------
#  Discrete-action Soft Actor–Critic for CartPole-v1
#  – twin critics, categorical actor, auto-entropy tuning
#  – Brownian process noise injected into the state
# ------------------------------------------------------------
import os, math, random, time
from collections import deque, namedtuple
from itertools import count

import gymnasium as gym
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
from tqdm import trange

# ------------------------------------------------------------
#  Helper: physics time-step  (CartPole ≈ 0.02 s)
# ------------------------------------------------------------
def _default_dt(env, fallback=0.02):
    return getattr(env.unwrapped, "tau", fallback)

# ------------------------------------------------------------
#  Hyper-parameters
# ------------------------------------------------------------
env_id = "CartPole-v1"
env    = gym.make(env_id)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class Args:
    # RL settings -----------------------------------------------------------
    gamma           = 0.99
    lr              = 3e-4
    batch_size      = 256
    replay_size     = 50_000
    min_buffer      = 1_000
    episodes        = 400
    steps_per_ep    = 500               # CartPole terminates earlier if solved
    tau             = 0.005             # Polyak coefficient for target critic
    # Entropy/temperature ---------------------------------------------------
    target_entropy  = -env.action_space.n        # ≈ −|A|
    alpha_lr        = 3e-4
    # Brownian noise --------------------------------------------------------
    sigma           = 0.1               # diffusion coefficient  σ
    dt              = _default_dt(env)
    # Evaluation ------------------------------------------------------------
    eval_traj_len   = 500
    eval_num_traj   = 10
    # Misc ------------------------------------------------------------------
    device          = device
args = Args()

# ------------------------------------------------------------
#  Replay buffer
# ------------------------------------------------------------
Transition = namedtuple(
    "Transition", ("state", "action", "reward", "next_state", "done")
)

class ReplayBuffer:
    def __init__(self, capacity: int):
        self.buf = deque(maxlen=capacity)

    def push(self, *args):
        self.buf.append(Transition(*args))

    def sample(self, n: int):
        batch  = random.sample(self.buf, n)
        trans  = Transition(*zip(*batch))
        to_t   = lambda x: torch.as_tensor(x, device=args.device,
                                           dtype=torch.float32)
        s      = to_t(trans.state)
        a      = torch.as_tensor(trans.action, device=args.device).long()
        r      = to_t(trans.reward).unsqueeze(1)
        s2     = to_t(trans.next_state)
        d      = to_t(trans.done).unsqueeze(1)
        return s, a, r, s2, d

    def __len__(self):
        return len(self.buf)

memory = ReplayBuffer(args.replay_size)

# ------------------------------------------------------------
#  Networks
# ------------------------------------------------------------
def mlp(in_dim: int, out_dim: int, hidden: int = 128):
    """Two-layer MLP."""
    return nn.Sequential(
        nn.Linear(in_dim, hidden), nn.ReLU(),
        nn.Linear(hidden, hidden), nn.ReLU(),
        nn.Linear(hidden, out_dim)
    )

class Critic(nn.Module):
    """Twin Q-networks:  Q₁(s,a), Q₂(s,a) for *each* discrete action."""
    def __init__(self, s_dim: int, nA: int):
        super().__init__()
        self.q1 = mlp(s_dim, nA)
        self.q2 = mlp(s_dim, nA)

    def forward(self, s):
        return self.q1(s), self.q2(s)     # each (B, nA)

class Actor(nn.Module):
    """
    Categorical policy  π(·|s).
    · during interaction: Gumbel-Max sampling (no grad needed)
    · during back-prop   : uses full distribution for gradients
    """
    def __init__(self, s_dim: int, nA: int):
        super().__init__()
        self.net = mlp(s_dim, nA)

    def forward(self, s, *, deterministic=False, sample=True):
        logits = self.net(s)                         # (B, nA)
        logp   = F.log_softmax(logits, dim=-1)
        p      = logp.exp()

        if deterministic:                            # argmax action
            a = torch.argmax(p, dim=-1)
            return a, logp, p

        if sample:                                   # Gumbel-Max
            gumbel = -torch.empty_like(logp).exponential_().log()
            a = torch.argmax(logp + gumbel, dim=-1)
            return a, logp, p

        # when gradients are needed but no sample
        return logp, p

# Instantiate nets & optimisers --------------------------------------------
nS, nA = env.observation_space.shape[0], env.action_space.n
actor   = Actor(nS, nA).to(device)
critic  = Critic(nS, nA).to(device)
target  = Critic(nS, nA).to(device)
target.load_state_dict(critic.state_dict())

actor_opt  = torch.optim.Adam(actor.parameters(),  lr=args.lr)
critic_opt = torch.optim.Adam(critic.parameters(), lr=args.lr)

log_alpha  = torch.tensor(0.0, requires_grad=True, device=device)
alpha_opt  = torch.optim.Adam([log_alpha], lr=args.alpha_lr)

# ------------------------------------------------------------
#  Evaluation (greedy argmax)
# ------------------------------------------------------------
@torch.no_grad()
def evaluate_policy(actor, args, env):
    actor.eval()
    returns = []
    for _ in range(args.eval_num_traj):
        obs, _ = env.reset()
        ep_ret = 0.0
        for t in range(args.eval_traj_len):
            s = torch.as_tensor(obs, device=device, dtype=torch.float32)
            a, _, _ = actor(s.unsqueeze(0), deterministic=True)
            a_idx   = int(a.item())
            obs2, r, term, trunc, _ = env.step(a_idx)
            ep_ret += r * (args.gamma ** t)

            # Brownian state perturbation
            st   = torch.as_tensor(obs2, device=device)
            dW   = torch.randn_like(st) * math.sqrt(args.dt)
            st_n = st + args.sigma * dW
            env.unwrapped.state = st_n.cpu().numpy().copy()
            obs = st_n.cpu().numpy()

            if term or trunc:
                break
        returns.append(ep_ret)
    actor.train()
    return np.mean(returns)

# ------------------------------------------------------------
#  Training loop
# ------------------------------------------------------------
reward_history = []
for ep in trange(args.episodes, desc="Episodes"):
    obs, _ = env.reset()
    for t in range(args.steps_per_ep):
        s_tensor = torch.as_tensor(obs, device=device, dtype=torch.float32)

        # ---------- choose action (stochastic) ----------
        a_idx, _, _ = actor(s_tensor.unsqueeze(0))     # Gumbel-Max sample
        a_idx = int(a_idx.item())

        obs2, r, term, trunc, _ = env.step(a_idx)

        # ---------- Brownian noise injection ----------
        st   = torch.as_tensor(obs2, device=device)
        dW   = torch.randn_like(st) * math.sqrt(args.dt)
        st_n = st + args.sigma * dW
        env.unwrapped.state = st_n.cpu().numpy().copy()

        done = term or trunc
        memory.push(obs, a_idx, r, st_n.cpu().numpy(), float(done))
        obs  = st_n.cpu().numpy()

        # ---------- update networks ----------
        # ---------- update networks ----------
        if len(memory) >= args.min_buffer:
            s_b, a_b, r_b, s2_b, d_b = memory.sample(args.batch_size)
        
            # =====================================================
            # 1) Critic update  (same as before)
            # =====================================================
            with torch.no_grad():
                # target V(s')
                logp2, p2 = actor(s2_b, sample=False)            # no graph here
                q1_t, q2_t = target(s2_b)
                q_min_next = torch.min(q1_t, q2_t)
                alpha      = log_alpha.exp()
                v_next     = (p2 * (q_min_next - alpha * logp2)).sum(-1, keepdim=True)
                y          = r_b + args.gamma * (1 - d_b) * v_next
        
            q1, q2  = critic(s_b)                                # each (B,nA)
            q1_sa   = q1.gather(1, a_b.unsqueeze(1))
            q2_sa   = q2.gather(1, a_b.unsqueeze(1))
            critic_loss = F.mse_loss(q1_sa, y) + F.mse_loss(q2_sa, y)
        
            critic_opt.zero_grad()
            critic_loss.backward()
            critic_opt.step()
        
            # =====================================================
            # 2) Actor update  (fresh forward pass!)
            # =====================================================
            logp_pi, p_pi = actor(s_b, sample=False)             # graph for actor
            with torch.no_grad():                                # critic output is a constant
                q1_pi, q2_pi = critic(s_b)
                q_pi = torch.min(q1_pi, q2_pi)                   # detach from critic
        
            alpha = log_alpha.exp().detach()                     # treat α as const here
            actor_loss = ((alpha * logp_pi) - q_pi) * p_pi       # (B,nA)
            actor_loss = actor_loss.sum(-1).mean()
        
            actor_opt.zero_grad()
            actor_loss.backward()
            actor_opt.step()
        
            # =====================================================
            # 3) Temperature (alpha) update – *no gradients* w.r.t actor
            # =====================================================
            with torch.no_grad():
                # reuse the policy outputs we just computed, but detach them
                entropy = -(logp_pi * p_pi).sum(-1, keepdim=True)  # H(π(s))
            alpha_loss = (log_alpha * (entropy - args.target_entropy)).mean()
        
            alpha_opt.zero_grad()
            alpha_loss.backward()      # only log_alpha has gradients
            alpha_opt.step()
        
            # =====================================================
            # 4) Soft-update target critic
            # =====================================================
            with torch.no_grad():
                for p, tp in zip(critic.parameters(), target.parameters()):
                    tp.data.lerp_(p.data, args.tau)


        if done:
            break

    # ------------ evaluation / logging ---------------
    avg_ret = evaluate_policy(actor, args, env)
    reward_history.append(avg_ret)
    if (ep + 1) % 10 == 0:
        print(f"[Ep {ep+1:4d}]  Eval return = {avg_ret:7.2f}   α = {log_alpha.exp():.3f}")

env.close()

# ------------------------------------------------------------
#  Simple plot
# ------------------------------------------------------------
plt.figure(figsize=(7,4))
plt.plot(reward_history, label="Eval return")
plt.axhline(475, ls="--", c="k", alpha=0.5, label="Solving threshold")
mean100 = np.convolve(reward_history, np.ones(100)/100, mode="valid")
if len(mean100):
    plt.plot(range(99, 99 + len(mean100)), mean100, lw=2, label="100-ep mean")
plt.xlabel("Episode"); plt.ylabel("Discounted return")
plt.legend(); plt.tight_layout()
plt.show()
```
